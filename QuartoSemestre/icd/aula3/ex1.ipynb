```python
import numpy as np
import matplotlib.pyplot as plt
from sklearn.linear_model import LinearRegression

N = 10
x = np.linspace(0,10,N)
y = x**3 + np.random.normal(0,50, N)
plt.plot(x,y, 'ro')
plt.ylabel("y", fontsize=15)
plt.xlabel("x", fontsize=15)

plt.show(True)
```

```python
from sklearn.preprocessing import PolynomialFeatures

# define a transformação nos dados
transformer_deg3 = PolynomialFeatures(degree=3, include_bias=False)
transformer_deg5 = PolynomialFeatures(degree=5, include_bias=False)
x = x.reshape((-1, 1))
transformer_deg3.fit(x)
transformer_deg5.fit(x)

# transforma os dados incluindo uma nova coluna com valores quadráticos
x_3 = transformer_deg3.transform(x)
x_5 = transformer_deg5.transform(x)
```

```python
model_3 = LinearRegression().fit(x_3, y)
y_pred_3 = model_3.predict(x_3)

model_5 = LinearRegression().fit(x_5, y)
y_pred_5 = model_5.predict(x_5)

plt.plot(x,y, 'ro', label='Dados originais')
plt.plot(x,y_pred_3, 'bo-',color='green', label = 'Dados preditos para P_3')
plt.plot(x,y_pred_5, 'bo-', label = 'Dados preditos para P_5')
plt.ylabel("y", fontsize=15)
plt.xlabel("x", fontsize=15)
plt.legend()
plt.show(True)
```

```python
from sklearn.metrics import r2_score
R2_deg3 = r2_score(y, y_pred_3)
R2_deg5 = r2_score(y, y_pred_5)

print('R2 para polinomio de grau 3: ', R2_deg3)
print('R2 para polinomio de grau 5: ', R2_deg5)

if (R2_deg3 > R2_deg5):
    print('O polinomio de grau 3 se ajusta melhor aos dados')
else:
    print('O polinomio de grau 5 se ajusta melhor aos dados')
```
